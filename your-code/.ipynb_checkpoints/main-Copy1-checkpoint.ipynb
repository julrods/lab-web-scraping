####  20 latest earthquakes info (date, time, latitude, longitude and region name) by the EMSC as a pandas dataframe

````python
# This is the url you will scrape in this exercise
url = 'https://www.emsc-csem.org/Earthquake/'
````

````python
resp = requests.get(url, headers=headers)
````

````python
resp.status_code
````

````python
soup = BeautifulSoup(resp.text, 'html.parser')
````

````python
earthquakes = soup.findAll('tr', attrs={"class":"ligne1 normal" or "ligne2 normal"})[:20]
````

````python
earthquakes_dict = [{'date':earthquake.find('td', attrs={"class":"tabev6"}).find('b').find('a').text.split(u'\xa0\xa0\xa0')[0],
                     'time':earthquake.find('td', attrs={"class":"tabev6"}).find('b').find('a').text.split(u'\xa0\xa0\xa0')[1],
                     'latitude':earthquake.findAll('td', attrs={"class":"tabev1"})[0].text.strip(),
                     'longitude':earthquake.findAll('td', attrs={"class":"tabev1"})[1].text.strip(),
                     'region':earthquake.find('td', attrs={"class":"tb_region"}).text.strip()}
                   for earthquake in earthquakes]
````

````python
df = pd.DataFrame(earthquakes_dict)
````

````python
df
````

#### Top 10 languages by number of native speakers stored in a Pandas Dataframe

````python
# This is the url you will scrape in this exercise
url = 'https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers'
````

````python
resp = requests.get(url, headers=headers)
````

````python
resp.status_code
````

````python
html = resp.text
````

````python
soup = BeautifulSoup(html, 'html.parser')
````

````python
soup
````

````python
table = soup.findAll('table')[1]
````

````python
rows = table.find('tbody').findAll('tr')[1:]
````

````python
languages_dict = [{'language':row.findAll('td')[1].text.strip(),
                  'millions_speakers':row.findAll('td')[2].text.strip()}
                 for row in rows[:10]]
````

````python
df = pd.DataFrame(languages_dict)
````

````python
df
````
